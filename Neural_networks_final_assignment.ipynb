**PART A**



Task description:
In the PART A we need to build a baseline model.
Using the Keras library to build a neural network with the following parameters:
- **One hidden layer** of **10 nodes**, and a **ReLU** activation function
- Use the **adam** optimizer and the **mean squared error **as the loss function.

```
import keras
```

Let's upload our data:

```
import pandas as pd
import numpy as np
```

```
concrete_data = pd.read_csv('https://cocl.us/concrete_data')
concrete_data.head() # to show our data set
```

We need to separate our data on predictors and target

```
concrete_data_columns = concrete_data.columns
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']]
target = concrete_data['Strength']

target.head() # now we have two data set, "target" and "predictors", let's see what "target" dataset consists of
```

Now we need to split our data like that:
**30% data for testing** and the rest for training.

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.33)

X_train.head() # now we have 4 dataset and I show for example how "X_train" dataset looks like:
```

Now we need to create our regression model, let's create a separate function for that:

```
import keras 

from keras.models import Sequential
from keras.layers import Dense
```

```
data_shape = predictors.shape[1]
data_shape # to find out how much inputs should be in our model
```

The model, as task says, should consist of:
* **one hidden layer** of **10** nodes and **ReLU** activation function
* we should use **adam** optimizer and the **mean squared error** as the loss function

```
def our_model():
  # create model
  model = Sequential()
  model.add(Dense(10, activation='relu', input_shape = (data_shape,))) # hidden layer with 10 nodes and ReLU as activation function)
  model.add(Dense(1)) # the output layer

  # compile model 
  model.compile(optimizer='adam', loss='mean_squared_error')
  return model
```

Now we should train our model using 50 epochs

```
# build the model
model = our_model()
# fit the model
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, verbose=2)
```

Now we need to **evaluate our model** on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength.
We will use the **mean_squared_error** function from Scikit-learn.

```
from sklearn.metrics import mean_squared_error

true_data = target # here is our actual data, from original data set
predicted_data = model.predict(predictors) # getting predicted data
print(predicted_data)
print(predicted_data.shape)
print(true_data)
print(true_data.shape)

mse = mean_squared_error(true_data, predicted_data, squared=True) # mean squared error between actual and predicted data
mse
```

Now the task is to repeat all the steps 50 times and get a list of 50 mean squared errors.
Let me remind the steps we did:
1. Randomly split the data into a training and test sets by holding 30% of the data for testing. 
2. Train the model on the training data using 50 epochs
3. Evaluate the model on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength.

Let's do it!

```
model = our_model()
true_data = target # actual data
mse_std_list = [] # initialize an empty list for 50 mean squared errors
for i in range(0,50):
  X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=24)
  model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, verbose=2) # train the model
  
  predicted_data = model.predict(predictors) # getting predicted data, using our trained model
  mse = mean_squared_error(true_data, predicted_data, squared=True) # find out mean squared error for the first iteration
  mse_std_list.append(mse) # add new mean squared error to our list for 50 of them

print(mse_std_list)
```

So the conclusion here, in the *1 iteration* the **mean squared error** was **162.8**, but *after* our model learnt during 50 iterations, the **mse** value became **37.54** which is much less compare with 162.8 at the first iteration

```
print(mse_std_list)
print(np.shape(mse_std_list)) # double check how much mean squared errors I have in my list
```

```
print('Standart deviation of mse is ' + str(np.std(mse_std_list)))
```

```
print('The mean of mean squared errors is ' + str(np.mean(mse_std_list)))
```

**Conclusion** for the **PART A**:
We can see, that in the first iteration our MSE was 1479.85. After our model was learning during 50 iterations, the MSE became 61.45.
The **mean** of MSEs (all 50) is 101.9
The **standart deviation** is 217.09.

**PART "B"**

In part "B" we are required to **normalize** our data.
By *subtracting the mean* from the individual predictors and *dividing by the standard deviation*.



```
predictors_norm = (predictors - predictors.mean()) / predictors.std() # normalizing of predictors
predictors_norm.head()
```

And now we should repeat the whole part "A" but with normalized data.

```
norm_model = our_model() # creating new model
true_data = target # actual data
mse_list2 = [] # empty list for the list of 50 mean squared errors

for i in range(0,50):
  X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.3, random_state=24)
  # the only different with the PART A is that we use "predictors_norm" set as predictors data
  norm_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, verbose=2)
  
  predicted_data = norm_model.predict(predictors_norm)
  mse2 = mean_squared_error(true_data, predicted_data, squared=True)
  mse_list2.append(mse2)

print(mse_list2)

print('Standart deviation of MSE is ' + str(np.std(mse_list2)))
print('The mean of mean squared errors is ' + str(np.mean(mse_list2)))
```

**Conclusion** for **PART B**.
In the *first iteration* the MSE was **410.82**.
After learning during *50 iterations*), *MSE* became **35.00**
The **standart deviation** of mse is **56.44**
and the **mean** of *MSE* is **51**

So, the result for normalize data is much better compare with non normalize data

* the first iteration, MSE was **1479.85**.

**PART C**

In part "C" we are required to increase the number of epochs to 100. And repeat part B after that (with 100 epochs).
All we need to change is number of epochs during our training.

**PART D**

Here we need to increase the number of hidden layers:
- Three hidden layers, each of 10 nodes and ReLU activation function.

and repeat part B. Small reminder here, in part B we train our model only during 50 epochs, but using the normalized data.

Let's code!

```
def layers3_model():
  # create model
  model = Sequential()
  model.add(Dense(10, activation='relu', input_shape = (data_shape,))) # firts hidden layer
  model.add(Dense(10, activation='relu')) # second hidden layer
  model.add(Dense(10, activation='relu')) # third hidden layer
  model.add(Dense(1)) # output layer

  # compile model 
  model.compile(optimizer='adam', loss='mean_squared_error')
  return model


more_layers_model = layers3_model()
true_data = target #actual data
mse_list4 = [] # empty list for the list of 50 mean squared errors

for i in range(0,50):
  X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.3, random_state=24)
  more_layers_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, verbose=2)
  
  predicted_data = more_layers_model.predict(predictors_norm)
  mse4 = mean_squared_error(true_data, predicted_data, squared=True)
  mse_list4.append(mse4) # adding current MSE to the final list of 50 MSEs

print(mse_list4)

print('Standart deviation of MSEs is ' + str(np.std(mse_list4)))
print('The mean of mean squared errors is ' + str(np.mean(mse_list4)))
```
